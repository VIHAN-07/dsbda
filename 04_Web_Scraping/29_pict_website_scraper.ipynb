```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# PICT college website
url = "https://www.pict.edu.in"

try:
    response = requests.get(url, timeout=5)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Extract all text content (simplified)
    data = []
    for p in soup.find_all("p")[:10]:
        text = p.get_text(strip=True)
        if text:
            data.append({"content": text})
    
    # Extract all links
    for link in soup.find_all("a")[:10]:
        href = link.get("href", "")
        text = link.get_text(strip=True)
        if text:
            data.append({"link": href, "text": text})
    
    df = pd.DataFrame(data)
    print(df.head(10))
    
except Exception as e:
    print(f"Error: {e}")
    print("Note: Website may require special headers or may block requests")
```
