```python
import requests
from bs4 import BeautifulSoup

url="https://webscraper.io/test-sites/e-commerce/static/computers/laptops"

r=requests.get(url)

soup=BeautifulSoup(r.text,"html.parser")

data=soup.find_all("div",class_="thumbnail")

for i in data[:5]:

    print("Name :",i.a.text)

    print("Price :",i.h4.text)

    print("Review :",i.find("p",class_="description").text)

    print("Rating :",len(i.find_all("span",class_="glyphicon-star")))

    print("Comments :",i.find_all("p")[-1].text)

    print("-"*50)
```
