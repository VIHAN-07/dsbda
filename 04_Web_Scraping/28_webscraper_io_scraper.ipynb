```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch webscraper.io e-commerce test site
url = "https://webscraper.io/test-sites/e-commerce/static"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Extract products (simplified)
products = []
for product in soup.find_all("div", class_="col-sm-4"):
    name = product.find("h4", class_="pull-right price")
    description = product.find("p", class_="description")
    rating = product.find("p", class_="pull-right")
    
    products.append({
        "name": name.text if name else "",
        "description": description.text if description else "",
        "rating": rating.text if rating else "",
    })

df = pd.DataFrame(products)
print(df.head(10))
```
