```python
import requests
import json
import pandas as pd

# Steam API endpoint for CS:GO reviews (app 730)
url = "https://store.steampowered.com/appreviews/730?json=1"

# Fetch reviews
response = requests.get(url)
data = response.json()

# Extract reviews
reviews = []
for review in data["reviews"]:
    reviews.append({
        "author": review.get("author", {}).get("steamid", ""),
        "rating": review.get("voted_up", 0),
        "comment": review.get("review", ""),
        "helpful": review.get("votes_helpful", 0),
        "funny": review.get("votes_funny", 0),
    })

df = pd.DataFrame(reviews)
print(df.head(10))
```
