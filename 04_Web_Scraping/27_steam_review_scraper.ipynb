```python
import requests

url = "https://store.steampowered.com/appreviews/730?json=1"

r = requests.get(url)

data = r.json()

reviews = data["reviews"]

for i in reviews[:5]:

    name = i["author"]["steamid"]

    review = i["review"]

    rating = i["voted_up"]

    print("Customer :",name)

    print("Review :",review)

    print("Rating :",rating)

    print("-"*50)
```
