```python
import pandas as pd

# load dataset
df = pd.read_csv("datasets/forestfires.csv")

# a) create subsets
not_affected = df[df["area"] == 0]

partial = df[
    (df["area"] > 0) &
    (df["area"] <= 10)
]

mostly = df[df["area"] > 10]

print(not_affected.head())


# b) merge subsets
merged = pd.concat([not_affected, partial])

print(merged.head())


# c) sort data
sorted_df = df.sort_values(
    ["temp", "wind", "area"]
)

print(sorted_df.head())


# d) transpose data
print(df.head().T)


# e) melt data
melted = pd.melt(
    df,
    value_vars=["temp", "wind", "area"]
)

print(melted.head())


# f) wide format
wide = melted.pivot_table(
    columns="variable",
    values="value"
)

print(wide)
```
