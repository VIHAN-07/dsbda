```python
import pandas as pd

# Load dataset

df = pd.read_csv("DSBDALExam DataSets/forestfires/forestfires.csv")

# a) subsets by area class
max_area = df["area"].max()
df["area_class"] = pd.cut(
    df["area"],
    bins=[-0.1, 0, 10, max_area],
    labels=["NotAffected", "Partially affected", "Mostly affected"],
)

not_affected = df[df["area_class"] == "NotAffected"]
partially = df[df["area_class"] == "Partially affected"]
mostly = df[df["area_class"] == "Mostly affected"]

# b) merge two subsets
merged = pd.concat([not_affected, partially], ignore_index=True)

# c) sort by temperature, wind, area
sorted_df = df.sort_values(["temp", "wind", "area"])

# d) transpose (small sample)
transposed = df.head(5).T

# e) melt to long format
melted = pd.melt(
    df,
    id_vars=["area_class"],
    value_vars=["temp", "wind", "area"],
    var_name="metric",
    value_name="value",
)

# f) cast to wide format (pivot)
wide = melted.pivot_table(
    index="area_class",
    columns="metric",
    values="value",
    aggfunc="mean",
).reset_index()

not_affected.head()
```
