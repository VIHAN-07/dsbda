```python
import pandas as pd

cols = [
    "sepal_length",
    "sepal_width",
    "petal_length",
    "petal_width",
    "species",
]

df = pd.read_csv("DSBDALExam DataSets/Iris/Iris.csv", header=None, names=cols)

# g) subsets by species
setosa = df[df["species"] == "Iris-setosa"]
versicolor = df[df["species"] == "Iris-versicolor"]

# h) merge two subsets
merged = pd.concat([setosa, versicolor], ignore_index=True)

# i) sort by petal length
sorted_df = df.sort_values("petal_length")

# j) transpose (small sample)
transposed = df.head(5).T

# k) melt to long format
melted = pd.melt(
    df,
    id_vars=["species"],
    value_vars=["sepal_length", "sepal_width"],
    var_name="measure",
    value_name="value",
)

# l) cast to wide format (pivot)
wide = melted.pivot_table(
    index="species",
    columns="measure",
    values="value",
    aggfunc="mean",
).reset_index()

setosa.head()
```
