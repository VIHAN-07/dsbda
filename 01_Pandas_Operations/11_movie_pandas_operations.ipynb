```python
import pandas as pd

df = pd.read_csv("DSBDAL_datasets/tmdb_5000_movies.csv")

# m) subsets by original_language
en = df[df["original_language"] == "en"]
fr = df[df["original_language"] == "fr"]

# n) merge two subsets
merged = pd.concat([en, fr], ignore_index=True)

# o) sort by customer ratings (vote_average)
sorted_df = df.sort_values("vote_average", ascending=False)

# p) transpose (small sample)
transposed = df.head(5).T

# q) melt to long format
melted = pd.melt(
    df,
    id_vars=["original_language", "title"],
    value_vars=["vote_average", "vote_count"],
    var_name="metric",
    value_name="value",
)

# r) cast to wide format (pivot)
wide = melted.pivot_table(
    index="original_language",
    columns="metric",
    values="value",
    aggfunc="mean",
).reset_index()

en.head()
```
