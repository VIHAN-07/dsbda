```python
import pandas as pd

# Column names for UCI Hepatitis dataset
cols = [
    "class",
    "age",
    "sex",
    "steroid",
    "antivirals",
    "fatigue",
    "malaise",
    "anorexia",
    "liver_big",
    "liver_firm",
    "spleen_palpable",
    "spiders",
    "ascites",
    "varices",
    "bilirubin",
    "alk_phos",
    "sgot",
    "albumin",
    "protime",
    "histology",
]

df = pd.read_csv(
    "datasets/hepatitis.csv",
    header=None,
    names=cols,
    na_values="?",
)

# a) subsets by sex (1=male, 2=female)
male = df[df["sex"] == 1]
female = df[df["sex"] == 2]

# b) merge two subsets
merged = pd.concat([male, female], ignore_index=True)

# c) sort by age, SGOT, PROTIME
sorted_df = df.sort_values(["age", "sgot", "protime"])

# d) transpose (small sample)
transposed = df.head(5).T

# e) melt to long format
melted = pd.melt(
    df,
    id_vars=["sex"],
    value_vars=["age", "sgot", "protime"],
    var_name="metric",
    value_name="value"
)

# f) cast to wide format (pivot)
wide = melted.pivot_table(
    index="sex",
    columns="metric",
    values="value",
    aggfunc="mean"
).reset_index()

male.head()
```
