```python
import pandas as pd

df = pd.read_csv(
    "DSBDAL_datasets/Facebook_metrics/dataset_Facebook.csv",
    sep=";",
)

# a) subsets by Type
photo = df[df["Type"] == "Photo"]
video = df[df["Type"] == "Video"]

# b) merge two subsets
merged = pd.concat([photo, video], ignore_index=True)

# c) sort by Page total likes
sorted_df = df.sort_values("Page total likes")

# d) transpose (small sample)
transposed = df.head(5).T

# e) melt to long format
melted = pd.melt(
    df,
    id_vars=["Type"],
    value_vars=["Page total likes", "Lifetime Post Total Reach"],
    var_name="metric",
    value_name="value",
)

# f) cast to wide format (pivot)
wide = melted.pivot_table(
    index="Type",
    columns="metric",
    values="value",
    aggfunc="mean",
).reset_index()

photo.head()
```
