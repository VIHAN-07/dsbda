```python
import pandas as pd

# Load Facebook dataset
df = pd.read_csv(
    "datasets/dataset_Facebook.csv",
    sep=";"
)

# Display first 5 rows
print(df.head())

# ---------------------------------------------------
# a) Create data subsets for type of post
# ---------------------------------------------------

photo = df[df["Type"] == "Photo"]
video = df[df["Type"] == "Video"]

print("\nPhoto Posts:")
print(photo.head())

print("\nVideo Posts:")
print(video.head())

# ---------------------------------------------------
# b) Merge two subsets
# ---------------------------------------------------

merged = pd.concat([photo, video], ignore_index=True)

print("\nMerged Data:")
print(merged.head())

# ---------------------------------------------------
# c) Sort data on Page total likes
# ---------------------------------------------------

sorted_df = df.sort_values(by="Page total likes")

print("\nSorted Data:")
print(sorted_df.head())

# ---------------------------------------------------
# d) Transposing data
# ---------------------------------------------------

transposed = df.head(5).T

print("\nTransposed Data:")
print(transposed)

# ---------------------------------------------------
# e) Melting data to long format
# ---------------------------------------------------

melted = pd.melt(
    df,
    id_vars=["Type"],
    value_vars=[
        "Page total likes",
        "Lifetime Post Total Reach"
    ],
    var_name="Metric",
    value_name="Value"
)

print("\nMelted Data:")
print(melted.head())

# ---------------------------------------------------
# f) Casting data to wide format
# ---------------------------------------------------

wide = melted.pivot_table(
    index="Type",
    columns="Metric",
    values="Value",
    aggfunc="mean"
).reset_index()

print("\nWide Format Data:")
print(wide)
```
