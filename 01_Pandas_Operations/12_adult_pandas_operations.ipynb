```python
import pandas as pd

cols = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education-num",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital-gain",
    "capital-loss",
    "hours-per-week",
    "native-country",
    "income",
]

df = pd.read_csv(
    "DSBDAL_datasets/adult.csv",
    header=None,
    names=cols,
    skipinitialspace=True,
)

# s) subsets by country, sex, race
us = df[df["native-country"] == "United-States"]
male = df[df["sex"] == "Male"]
white = df[df["race"] == "White"]

# t) merge two subsets
merged = pd.concat([male, white], ignore_index=True)

# u) sort by "customer ratings" (not present) -> hours-per-week
sorted_df = df.sort_values("hours-per-week")

# v) transpose (small sample)
transposed = df.head(5).T

# w) melt to long format
melted = pd.melt(
    df,
    id_vars=["sex", "race"],
    value_vars=["age", "hours-per-week"],
    var_name="metric",
    value_name="value",
)

# x) cast to wide format (pivot)
wide = melted.pivot_table(
    index=["sex", "race"],
    columns="metric",
    values="value",
    aggfunc="mean",
).reset_index()

us.head()
```
