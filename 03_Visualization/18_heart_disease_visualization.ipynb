```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
cols = [
    "age","sex","cp","trestbps","chol",
    "fbs","restecg","thalach","exang",
    "oldpeak","slope","ca","thal","target"
]

df = pd.read_csv("datasets/Cleavland.csv", header=None, names=cols)


# a) Histogram
# Objective: To see age distribution

plt.hist(df["age"], edgecolor="black")
plt.title("Age Histogram")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()  


# b) Pie Chart
# Objective: To show proportion of heart disease classes

df["target"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Target Pie Chart")
plt.ylabel("")
plt.show()


# c) Box Plot
# Objective: To detect spread and outliers in cholesterol

plt.boxplot(df["chol"])
plt.title("Cholesterol Boxplot")
plt.ylabel("Chol")
plt.show()


# d) Scatter Plot
# Objective: To see relation between age and max heart rate

plt.scatter(df["age"], df["thalach"])
plt.title("Age vs Thalach")
plt.xlabel("Age")
plt.ylabel("Thalach")
plt.show()


# e) Add boxplots to scatterplot
# Objective: To compare age and cholesterol with boxplots

fig = plt.figure(figsize=(6,6))

# scatter plot
plt.subplot(2,2,3)
plt.scatter(df["age"], df["chol"])
plt.xlabel("Age")
plt.ylabel("Chol")

# age boxplot
plt.subplot(2,2,1)
plt.boxplot(df["age"], vert=False)

# cholesterol boxplot
plt.subplot(2,2,4)
plt.boxplot(df["chol"])

plt.show()
```
