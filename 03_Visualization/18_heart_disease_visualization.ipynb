```python
import pandas as pd
import matplotlib.pyplot as plt

cols = [
    "age",
    "sex",
    "cp",
    "trestbps",
    "chol",
    "fbs",
    "restecg",
    "thalach",
    "exang",
    "oldpeak",
    "slope",
    "ca",
    "thal",
    "target",
]

df = pd.read_csv("DSBDALExam DataSets/HeartDisease/Cleavland.csv", header=None, names=cols)

# Objective: See the distribution of age.
plt.figure()
plt.hist(df["age"], bins=10, edgecolor="black")
plt.title("Age distribution")
plt.xlabel("age")
plt.ylabel("count")
plt.show()

# Objective: Show proportion of target classes (disease vs no disease).
plt.figure()
target_counts = df["target"].value_counts().sort_index()
plt.pie(target_counts, labels=target_counts.index, autopct="%1.1f%%")
plt.title("Target class proportion")
plt.ylabel("")
plt.show()

# Objective: See spread and outliers in cholesterol.
plt.figure()
plt.boxplot(df["chol"])
plt.title("Cholesterol boxplot")
plt.ylabel("chol")
plt.show()

# Objective: See relationship between age and max heart rate.
plt.figure()
plt.scatter(df["age"], df["thalach"], alpha=0.5)
plt.title("Age vs Max heart rate")
plt.xlabel("age")
plt.ylabel("thalach")
plt.show()

# Objective: Add boxplots to a scatterplot (marginal boxplots).
fig = plt.figure(figsize=(6, 6))
gs = fig.add_gridspec(2, 2, width_ratios=(4, 1), height_ratios=(1, 4), wspace=0.05, hspace=0.05)
ax_scatter = fig.add_subplot(gs[1, 0])
ax_xbox = fig.add_subplot(gs[0, 0], sharex=ax_scatter)
ax_ybox = fig.add_subplot(gs[1, 1], sharey=ax_scatter)

ax_scatter.scatter(df["age"], df["chol"], alpha=0.5)
ax_scatter.set_xlabel("age")
ax_scatter.set_ylabel("chol")
ax_scatter.set_title("Age vs Chol with boxplots")

ax_xbox.boxplot(df["age"], vert=False)
ax_ybox.boxplot(df["chol"], vert=True)
ax_xbox.axis("off")
ax_ybox.axis("off")

plt.show()
```
