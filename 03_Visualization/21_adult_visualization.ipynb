```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
cols = [
    "age","workclass","fnlwgt","education",    
    "education-num","marital-status","occupation",
    "relationship","race","sex","capital-gain",
    "capital-loss","hours-per-week",
    "native-country","income"
]

df = pd.read_csv(
    "datasets/adult.csv",
    header=None,
    names=cols,
    skipinitialspace=True
)


# f) Histogram
# Objective: To see age distribution

plt.hist(df["age"], edgecolor="black")
plt.title("Age Histogram")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()


# g) Dot Plot
# Objective: To see individual hours-per-week values

plt.plot(df["hours-per-week"].head(100), "o")
plt.title("Dot Plot")
plt.xlabel("Rows")
plt.ylabel("Hours per week")
plt.show()


# h) Bar Plot
# Objective: To compare workclass counts

df["workclass"].value_counts().head(5).plot(kind="bar")

plt.title("Workclass Count")
plt.xlabel("Workclass")
plt.ylabel("Count")
plt.show()


# i) Line Chart
# Objective: To observe hours-per-week variation

plt.plot(df["hours-per-week"].head(50))

plt.title("Line Chart")
plt.xlabel("Rows")
plt.ylabel("Hours per week")
plt.show()


# j) Add Histogram and Scatter Plot to Boxplot
# Objective: To compare distribution and relation of age and hours-per-week

plt.figure(figsize=(12,4))

# Boxplot
plt.subplot(1,3,1)
plt.boxplot(df["hours-per-week"])
plt.title("Boxplot")

# Histogram
plt.subplot(1,3,2)
plt.hist(df["hours-per-week"])
plt.title("Histogram")

# Scatter Plot
plt.subplot(1,3,3)
plt.scatter(df["age"], df["hours-per-week"])

plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Hours per week")

plt.show()
```
