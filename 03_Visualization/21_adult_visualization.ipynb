```python
import pandas as pd
import matplotlib.pyplot as plt

cols = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education-num",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital-gain",
    "capital-loss",
    "hours-per-week",
    "native-country",
    "income",
]

df = pd.read_csv(
    "DSBDAL_datasets/adult.csv",
    header=None,
    names=cols,
    skipinitialspace=True,
)

# Objective: See the distribution of age.
plt.figure()
plt.hist(df["age"], bins=10, edgecolor="black")
plt.title("Age distribution")
plt.xlabel("age")
plt.ylabel("count")
plt.show()

# Objective: Show individual values of hours-per-week (dot plot).
plt.figure()
plt.plot(df["hours-per-week"].head(100), "o", alpha=0.5)
plt.title("Hours-per-week (dot plot, first 100 rows)")
plt.xlabel("row")
plt.ylabel("hours-per-week")
plt.show()

# Objective: Compare counts of top workclass categories.
plt.figure()
df["workclass"].value_counts().head(5).plot(kind="bar")
plt.title("Top workclass counts")
plt.xlabel("workclass")
plt.ylabel("count")
plt.show()

# Objective: See trend of hours-per-week across records.
plt.figure()
plt.plot(df["hours-per-week"].head(50))
plt.title("Hours-per-week (first 50 rows)")
plt.xlabel("row")
plt.ylabel("hours-per-week")
plt.show()

# Objective: Compare distribution/outliers with boxplot, histogram, and scatter.
fig, axes = plt.subplots(1, 3, figsize=(12, 3))
axes[0].boxplot(df["hours-per-week"])
axes[0].set_title("Hours-per-week boxplot")
axes[1].hist(df["hours-per-week"], bins=10, edgecolor="black")
axes[1].set_title("Hours-per-week histogram")
axes[2].scatter(df["age"], df["hours-per-week"], alpha=0.5)
axes[2].set_title("Age vs Hours-per-week")
axes[2].set_xlabel("age")
axes[2].set_ylabel("hours-per-week")
plt.tight_layout()
plt.show()
```
