```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
df = pd.read_csv("datasets/Cleavland.csv", header=None)

# column names
df.columns = [
    "age","sex","cp","trestbps","chol",
    "fbs","restecg","thalach","exang",
    "oldpeak","slope","ca","thal","target"
]

# a) Histogram
# Objective: To see distribution of age
plt.hist(df["age"] , edgecolor="black")
plt.title("Age Histogram")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()


# b) Dot Plot
# Objective: To see relation between age and cholesterol
plt.plot(df["age"], df["chol"], "o")
plt.title("Dot Plot")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.show()


# c) Bar Plot
# Objective: To compare chest pain type counts
df["cp"].value_counts().plot(kind="bar")
plt.title("Chest Pain Count")
plt.xlabel("CP Type")
plt.ylabel("Count")
plt.show()


# d) Line Chart
# Objective: To observe max heart rate variation
plt.plot(df["thalach"].head(50))
plt.title("Line Chart")
plt.xlabel("Rows")
plt.ylabel("Thalach")
plt.show()


# e) Add Histogram and Scatter Plot to Boxplot
# Objective: To compare cholesterol distribution and age relation

plt.figure(figsize=(12,4))

plt.subplot(131)
plt.boxplot(df["chol"])
plt.title("Boxplot")

plt.subplot(132)
plt.hist(df["chol"],edgecolor="black")
plt.title("Histogram")

plt.subplot(133)
plt.scatter(df["age"], df["chol"])
plt.title("Scatter")

plt.show()
```
