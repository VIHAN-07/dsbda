```python
import pandas as pd
import matplotlib.pyplot as plt

cols = [
    "age",
    "sex",
    "cp",
    "trestbps",
    "chol",
    "fbs",
    "restecg",
    "thalach",
    "exang",
    "oldpeak",
    "slope",
    "ca",
    "thal",
    "target",
]

df = pd.read_csv("DSBDALExam DataSets/HeartDisease/Cleavland.csv", header=None, names=cols)

# Objective: See the distribution of age.
plt.figure()
plt.hist(df["age"], bins=10, edgecolor="black")
plt.title("Age distribution")
plt.xlabel("age")
plt.ylabel("count")
plt.show()

# Objective: Observe relationship between age and cholesterol (dot plot).
plt.figure()
plt.plot(df["age"], df["chol"], "o", alpha=0.5)
plt.title("Age vs Cholesterol (dot plot)")
plt.xlabel("age")
plt.ylabel("chol")
plt.show()

# Objective: Compare counts of chest pain types.
plt.figure()
df["cp"].value_counts().sort_index().plot(kind="bar")
plt.title("Chest pain type counts")
plt.xlabel("cp")
plt.ylabel("count")
plt.show()

# Objective: See how max heart rate varies across samples.
plt.figure()
plt.plot(df["thalach"].head(50))
plt.title("Max heart rate (first 50 rows)")
plt.xlabel("row")
plt.ylabel("thalach")
plt.show()

# Objective: Compare distribution/outliers with boxplot, histogram, and scatter.
fig, axes = plt.subplots(1, 3, figsize=(12, 3))
axes[0].boxplot(df["chol"])
axes[0].set_title("Cholesterol boxplot")
axes[1].hist(df["chol"], bins=10, edgecolor="black")
axes[1].set_title("Cholesterol histogram")
axes[2].scatter(df["age"], df["chol"], alpha=0.5)
axes[2].set_title("Age vs Chol")
axes[2].set_xlabel("age")
axes[2].set_ylabel("chol")
plt.tight_layout()
plt.show()
```
