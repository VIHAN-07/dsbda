```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
cols = [
    "age","workclass","fnlwgt","education",
    "education-num","marital-status","occupation",
    "relationship","race","sex","capital-gain",
    "capital-loss","hours-per-week",
    "native-country","income"
]

df = pd.read_csv(
    "datasets/adult.csv",
    header=None,
    names=cols,
    skipinitialspace=True
)


# a) Histogram
# Objective: To see age distribution

plt.hist(df["age"])

plt.title("Age Histogram")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()


# b) Pie Chart
# Objective: To show income proportion

df["income"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Income Pie Chart")
plt.ylabel("")

plt.show()


# c) Box Plot
# Objective: To see spread of hours-per-week by sex

df.boxplot(column="hours-per-week", by="sex")

plt.title("Hours per week by Sex")
plt.suptitle("")
plt.ylabel("Hours per week")

plt.show()


# d) Scatter Plot
# Objective: To see relation between age and hours-per-week

plt.scatter(df["age"], df["hours-per-week"])

plt.title("Age vs Hours per week")
plt.xlabel("Age")
plt.ylabel("Hours per week")

plt.show()


# e) Add boxplots to scatterplot
# Objective: To compare age and hours-per-week with boxplots

fig = plt.figure(figsize=(6,6))

# scatter plot
plt.subplot(2,2,3)
plt.scatter(df["age"], df["hours-per-week"])
plt.xlabel("Age")
plt.ylabel("Hours per week")

# age boxplot
plt.subplot(2,2,1)
plt.boxplot(df["age"], vert=False)

# hours-per-week boxplot
plt.subplot(2,2,4)
plt.boxplot(df["hours-per-week"])

plt.show()
```
