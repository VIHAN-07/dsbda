```python
import pandas as pd
import matplotlib.pyplot as plt

cols = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education-num",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital-gain",
    "capital-loss",
    "hours-per-week",
    "native-country",
    "income",
]

df = pd.read_csv(
    "DSBDAL_datasets/adult.csv",
    header=None,
    names=cols,
    skipinitialspace=True,
)

# Objective: See the distribution of age.
plt.figure()
plt.hist(df["age"], bins=10, edgecolor="black")
plt.title("Age distribution")
plt.xlabel("age")
plt.ylabel("count")
plt.show()

# Objective: Show income proportion.
plt.figure()
income_counts = df["income"].value_counts()
plt.pie(income_counts, labels=income_counts.index, autopct="%1.1f%%")
plt.title("Income proportion")
plt.ylabel("")
plt.show()

# Objective: See spread of hours-per-week by sex.
plt.figure()
df.boxplot(column="hours-per-week", by="sex")
plt.title("Hours-per-week by sex")
plt.suptitle("")
plt.ylabel("hours-per-week")
plt.show()

# Objective: See relationship between age and hours-per-week.
plt.figure()
plt.scatter(df["age"], df["hours-per-week"], alpha=0.5)
plt.title("Age vs Hours-per-week")
plt.xlabel("age")
plt.ylabel("hours-per-week")
plt.show()

# Objective: Add boxplots to a scatterplot (marginal boxplots).
fig = plt.figure(figsize=(6, 6))
gs = fig.add_gridspec(2, 2, width_ratios=(4, 1), height_ratios=(1, 4), wspace=0.05, hspace=0.05)
ax_scatter = fig.add_subplot(gs[1, 0])
ax_xbox = fig.add_subplot(gs[0, 0], sharex=ax_scatter)
ax_ybox = fig.add_subplot(gs[1, 1], sharey=ax_scatter)

ax_scatter.scatter(df["age"], df["hours-per-week"], alpha=0.5)
ax_scatter.set_xlabel("age")
ax_scatter.set_ylabel("hours-per-week")
ax_scatter.set_title("Age vs Hours-per-week with boxplots")

ax_xbox.boxplot(df["age"], vert=False)
ax_ybox.boxplot(df["hours-per-week"], vert=True)
ax_xbox.axis("off")
ax_ybox.axis("off")

plt.show()
```
